```scala
// Create RDDs for shakespeare/all_verbs/verb_dict files
val shakespeare = sc.textFile("shakespeare.txt")

val all_verbs = sc.textFile("all_verbs.txt").collect()

val verb_dict = sc.textFile("verb_dict.txt")

// shakespeare.take(20).foreach(println)
// println("-------------")
// all_verbs.take(20).foreach(println)
// println("-------------")
// verb_dict.take(20).foreach(println)
```

```scala
// Remove empty lines of Shakespeare rdd
val rdd1 = shakespeare.filter(l => l.length>0)

// rdd1.take(20).foreach(println)
```

```scala
// Remove punctuations of Shakespeare rdd
val rdd2 = rdd1.map(x => x.replaceAll("""[\p{Punct}]""",""))

// rdd2.take(20).foreach(println)
```

```scala
// Split the words of Shakespeare
val rdd3 = rdd2.flatMap(line => line.split(" "))

// rdd3.take(20).foreach(println)
```

```scala
// Remove empty words of Shakespeare
val rdd4 = rdd3.filter(_.nonEmpty)

// rdd4.take(20).foreach(println)
```

```scala
// Transfer to lower case of Shakespeare
val rdd5 = rdd4.map(word => word.toLowerCase())

// rdd5.take(20).foreach(println)
```

```scala
// only keep verbs of Shakespeare using all_verbs
val rdd6 = rdd5.filter(word => all_verbs.contains(word))

// rdd6.take(100).foreach(println)
```

```scala
// Generate key/value pairs
val rdd7 = rdd6.map(word => (word,1))

// rdd7.take(100).foreach(println)
```

```scala
// Aggregate the word frequencies
val rdd8 = rdd7.reduceByKey((a,b) => (a+b))

// rdd8.take(100).foreach(println)
```

```scala
// Generate rdds for different forms of verbs based on dictonaies
val dict1 = verb_dict.map(x => (x.split(",")(1), x.split(",")(0)))
val dict2 = verb_dict.map(x => (x.split(",")(2), x.split(",")(0)))
val dict3 = verb_dict.map(x => (x.split(",")(3), x.split(",")(0)))
val dict4 = verb_dict.map(x => (x.split(",")(4), x.split(",")(0)))
val dict5 = verb_dict.map(x => (x.split(",")(5), x.split(",")(0)))

//val (dict1, dict2, dict3, dict4, dict5) = verb_dict.map(x => ((x.split(",")(1), x.split(",")(0)), (x.split(",")(2), x.split(",")(0)) , (x.split(",")(3), x.split(",")(0)), (x.split(",")(4), x.split(",")(0)), (x.split(",")(5), x.split(",")(0))))

// dict5.collect().foreach(println)

// println(dict1.count())
// println(dict2.count())
// println(dict3.count())
// println(dict4.count())
// println(dict5.count())
```

```scala
// Merge the seperated dictionaries
val merged_dict = dict1 ++ dict2 ++ dict3 ++ dict4 ++ dict5

// println(merged_dict.count())

// val sort = merged_dict.sortByKey(true)
// sort.foreach(println)
```

```scala
// Remove duplicates of dictionaries by keys
val distinct_dict = merged_dict.distinct()

val sort1 = distinct_dict.sortByKey(true)
sort1.foreach(println)

println(distinct_dict.count())
```

```scala
// Join the word frequencies with dictionaries
val join_res = distinct_dict.join(rdd8)

// join_res.take(50).foreach(println)
```

```scala
// Take the values of word-dictionaries pairs
val frequency = join_res.map(x => (x._2))

// println(frequency.count())
// frequency.collect().foreach(println)
```

```scala
// Aggregate the same words
val same_frequency = frequency.reduceByKey((a,b) => (a+b))
println(same_frequency.count())
same_frequency.collect().foreach(println)
                                           
```

```scala
// Sort verbs by frequency descending
val result = same_frequency.sortBy(_._2, false)

// print out the top 10 most frequently used verb in Shakespeare
result.take(10).foreach(println)
```
